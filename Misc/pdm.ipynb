# Reaction Scheme: Metal–Film–Solution Interfaces

## Reactions at the Metal–Film Interface (mf-interface)

**(R1)**  
$$
Fe + v_{Fe}^{3'} \rightarrow Fe_{ox} + v_{Fe} + \frac{8}{3}e^-    :  \qquad    [k_1]
$$

**(R2)**  
$$
Fe \rightarrow Fe_{ox} + \frac{4}{3}V_{O}^{\cdot\cdot} + \frac{8}{3}e^-       :  \qquad   [k_2]
$$

## Reactions at the Film–Solution Interface (fs-interface)

**(R3)**  
$$
Fe_{ox} \rightarrow Fe^{3+} + v_{Fe}^{3'} + \frac{1}{3}e^-            :  \qquad [k_3]
$$

**(R4)**  
$$
V_{O}^{\cdot\cdot} + H_2O \rightarrow O_{ox} + H^+              :  \qquad  [k_4]
$$

**(R5)**  
$$
Fe_3O_4 + 8H^+ \rightarrow 2Fe^{3+} + Fe^{2+} + 4H_2O           :  \qquad[k_5']
$$

**(tp)**  
$$
Fe_3O_4 + 8H^+ + h^+ \rightarrow 3Fe^{3+} + 4H_2O              :  \qquad [k_{tp}]
$$

# Derivation of Rate Constant Units and Current Densities

## General Formulas

**Faraday's Law** for current density:

$$j = n F \cdot \text{rate}$$

- $j$: current density [A/m²]  
- $n$: number of electrons transferred  
- $F$: Faraday constant $\approx 96485 \, \text{C/mol}$  
- rate: molar flux $[\text{mol/(m}^2 \cdot \text{s})]$

---

## Reactions at the Metal–Film Interface

### **(R1)**  
$$Fe + v_{Fe}^{3'} \rightarrow Fe_{ox} + v_{Fe} + \frac{8}{3}e^-   :  \qquad [k_1]$$

**Rate law:**  
$$\text{rate} = k_1 [Fe] [v_{Fe}^{3'}]$$

**Units of rate:**  
$$\left[ \text{mol} \cdot \text{m}^{-2} \cdot \text{s}^{-1} \right]$$

**Units of concentrations:**  
$$[Fe], [v_{Fe}^{3'}] = \left[ \text{mol} \cdot \text{m}^{-2} \right]$$

**Deriving units of $k_1$:**
$$[k_1] = \frac{\text{mol} \cdot \text{m}^{-2} \cdot \text{s}^{-1}}{\text{mol}^2 \cdot \text{m}^{-4}} = \frac{\text{m}^2}{\text{mol} \cdot \text{s}}$$

**Current density:**
$$j_1 = \frac{8}{3} F k_1 [Fe] [v_{Fe}^{3'}]$$

---

### **(R2)**  
$$Fe \rightarrow Fe_{ox} + \frac{4}{3}V_{O}^{\cdot\cdot} + \frac{8}{3}e^-  :  \qquad   [k_2]$$

**Rate law:**  
$$\text{rate} = k_2 [Fe]$$

**Deriving units of $k_2$:**
$$[k_2] = \frac{\text{mol} \cdot \text{m}^{-2} \cdot \text{s}^{-1}}{\text{mol} \cdot \text{m}^{-2}} = \frac{\text{m}}{\text{s}}$$

**Current density:**
$$j_2 = \frac{8}{3} F k_2 [Fe]$$

---

## Reactions at the Film–Solution Interface

### **(R3)**  
$$Fe_{ox} \rightarrow Fe^{3+} + v_{Fe}^{3'} + \frac{1}{3}e^- :  \qquad   [k_3]$$

**Rate law:**
$$\text{rate} = k_3 [Fe_{ox}]$$

**Units:**
$$[k_3] = \frac{\text{m}}{\text{s}}$$

**Current density:**
$$j_3 = \frac{1}{3} F k_3 [Fe_{ox}]$$

---

### **(R4)**  
$$V_{O}^{\cdot\cdot} + H_2O \rightarrow O_{ox} + H^+:  \qquad  [k_4]$$

**Rate law:**
$$\text{rate} = k_4 [V_O^{\bullet\bullet}]$$

**Units of $k_4$:**
$$[k_4] = \frac{\mathrm{mol}}{\mathrm{m}^2 \cdot \mathrm{s}} \cdot \left(\frac{1}{\mathrm{mol} \cdot \mathrm{m}^{-2}}\right) = \mathrm{m/s}$$

**Current density:**
- No electron transfer  
- $j_4 = 0$

---

### **(R5)**  
$$Fe_3O_4 + 8H^+ \rightarrow 2Fe^{3+} + Fe^{2+} + 4H_2O  :  \qquad   [k_5']$$

**Rate law:**
$$\text{rate} = k_5' [Fe_3O_4] [H^+]^8$$

**Units of $k_5'$:**
$$[k_5'] = \frac{\text{mol} \cdot \text{m}^{-2} \cdot \text{s}^{-1}}{\text{mol}^9 \cdot \text{m}^{-18}} = \frac{\text{m}^{16}}{\text{mol}^8 \cdot \text{s}}$$

**Current density:**
$$j_5 = 0 \times F k_5' [Fe_3O_4][H^+]^8  = 0$$

---

### **(tp)**  
$$Fe_3O_4 + 8H^+ + h^+ \rightarrow 3Fe^{3+} + 4H_2O :  \qquad   [k_{tp}]$$

**Rate law:**
$$\text{rate} = k_{tp} [Fe_3O_4] [H^+]^8 [h^+]$$

**Units of $k_{tp}$:**
$$[k_{tp}] = \frac{\text{mol} \cdot \text{m}^{-2} \cdot \text{s}^{-1}}{\text{mol}^{10} \cdot \text{m}^{-20}} = \frac{\text{m}^{18}}{\text{mol}^9 \cdot \text{s}}$$

**Current density:**
$$j_{tp} = 1 \times  F k_{tp} [Fe_3O_4] [H^+]^8 [h^+]$$

# Rate Equation Calculations

$$\hat{k}_{\text{R1}} = k_{\text{R1}}^0 \exp\left(\alpha_1 \frac{3F \phi_c}{RT} \hat{\phi}_{\text{mf}}\right) \quad [\text{m/s}]$$

$$\hat{k}_{\text{R2}} = k_{\text{R2}}^0 \exp\left(\alpha_2 \frac{2F \phi_c}{RT} \hat{\phi}_{\text{mf}}\right) \quad [\text{mol}/\text{m}^2 \cdot \text{s}]$$

$$\hat{k}_{\text{R3}} = k_{\text{R3}}^0 \exp\left(\alpha_3 \frac{(3 - \delta)F \phi_c}{RT} \hat{\phi}_{\text{fs}}\right) \quad [\text{mol}/\text{m}^2 \cdot \text{s}]$$

$$\hat{k}_{\text{R4}} = k_{\text{R4}}^0 \quad [\text{m/s}]$$

$$\hat{k}_{\text{R5}} = k_{\text{R5}}^0 (c_{H^+})^n \quad [\text{mol}/\text{m}^2 \cdot \text{s}]$$

$$\hat{k}_{\text{TP}} = \hat{k}_{\text{TP}}^0 \hat{c}_h \hat{c}_e \exp\left(\alpha_{\text{TP}} \frac{F \phi_c}{RT} \hat{\phi}_{\text{fs}}\right) \quad [-]$$

$$\hat{k}_{\text{O2}} = k_{\text{O2}}^0 \exp\left(\alpha_{\text{O2}} \frac{2F \phi_c}{RT} \left(\hat{\phi}_{\text{ext}} - \hat{\phi}_{\text{O2,eq}}\right)\right)$$
